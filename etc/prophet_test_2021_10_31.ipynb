# Prophet test

## outdoor category before and after the pandemic

### 1. outdoor category before pandemic

```python
import pandas as pd
from prophet import Prophet
```

```python
outdoor_before_pandemic = pd.read_csv('outdoor_prediction_before_pandemic.csv')
outdoor_before_pandemic.head()
```

```python
m = Prophet()
m.fit(outdoor_before_pandemic)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast, ylabel='Number of Tourists', xlabel = 'Year')
```

```python
m.plot_components(forecast)
```
