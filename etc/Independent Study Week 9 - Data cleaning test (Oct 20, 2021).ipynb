## 2. Monthly Total by Category

only select the total number of tourists in each spot

```python
data_total = data_no_korean[(data_no_korean.foreigner_local == 'Total')]
data_total.head()
```

Select the data from '2016-Jan' to '2021-Mar'

```python
monthly_data_total = data_total[['tourist_spot_eng', 'category', '2016-Jan', '2016-Feb', '2016-Mar', '2016-Apr', '2016-May', '2016-Jun', '2016-Jul', '2016-Aug', '2016-Sep', '2016-Oct', '2016-Nov', '2016-Dec', '2017-Jan', '2017-Feb', '2017-Mar', '2017-Apr', '2017-May', '2017-Jun', '2017-Jul', '2017-Aug', '2017-Sep', '2017-Oct', '2017-Nov', '2017-Dec', '2018-Jan', '2018-Feb', '2018-Mar', '2018-Apr', '2018-May', '2018-Jun', '2018-Jul', '2018-Aug', '2018-Sep', '2018-Oct', '2018-Nov', '2018-Dec', '2019-Jan', '2019-Feb', '2019-Mar', '2019-Apr', '2019-May', '2019-Jun', '2019-Jul', '2019-Aug', '2019-Sep', '2019-Oct', '2019-Nov', '2019-Dec', '2020-Jan', '2020-Feb', '2020-Mar', '2020-Apr', '2020-May', '2020-Jun', '2020-Jul', '2020-Aug', '2020-Sep', '2020-Oct', '2020-Nov', '2020-Dec', '2021-Jan', '2021-Feb', '2021-Mar']]
monthly_data_total.head()
```

```python
monthly_data_total_agg = monthly_data_total.groupby('category').agg('sum')
monthly_data_total_agg
```

```python
monthly_data_total_agg_trans = monthly_data_total_agg.transpose()
monthly_data_total_agg_trans
```

```python
ax = monthly_data_total_agg_trans.plot(figsize = (15,6), kind='line', title='<2016-2021.March Monthly Total Number of Tourists by Category>')
ax.axvline(x='2020-Jan', label='COVID-19');
```

**(Problem)**

```python
monthly_data_total_agg_trans.reset_index(inplace=True)
monthly_data_total_agg_trans
```

```python
monthly_data_total_agg_trans_melt = monthly_data_total_agg_trans.melt(id_vars=['index'])
monthly_data_total_agg_trans_melt
```

- It clearly shows a significant decline in tourists in both outdoor and indoor tourist spots after the first confirmed COVID-19 case in January 2020..

## 5. 2016-2020 Outdoor and Indoor Proportion by Sub_Category

**1. Tourist spots in Nature sub_cagegory**
  - 2020-June: The proportion of nature tourist spots has increased compared to the previous year. It seems that the pandemic has made people prefer outdoor attractions.
  - I will update the data for 2021 so that we can see if people's preferences hold up or go back in time.

**2. Tourist spots in Social sub_cagegory**
  - Among five sub_categories, tourist spots in social sub_category experienced the largest decline in the number of tourists right after the outbreak of pandemic. However, from 2021 January, the proportion began to increase.
  - Tourists prefer more outdoor 'social' attractions than before COVID-19, and this trend is getting stronger.
